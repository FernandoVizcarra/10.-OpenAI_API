```python
from dotenv import load_dotenv
import os
load_dotenv()
api_key = os.getenv("TU_CLAVE_SECRETA_2")
```

```python
from openai import OpenAI  
import pandas as pd

# Creamos el cliente con tu clave secreta
client = OpenAI(api_key=api_key)

# Ejemplo de uso con los 7 productos
df_productos = pd.DataFrame({
    "Producto": ["Ensure",
                 "Vitamina C 1g efervescente",
                 "Omega 3",
                 "Melatonina 3-5mg",
                 "Eucerin Hyaluron-Filler",
                 "Retinol B3 serum",
                 "Biotina (Vit. H o B7)"]
})

resultados = []

# Iteramos sobre cada producto
for prod in df_productos["Producto"]:
    # Enviamos una pregunta al modelo GPT para generar el copy
    respuesta = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{
            "role": "user",
            "content": f"""Genera un copy estilo publicitario
            para el producto de farmacia '{prod}'
            pensado para usarse en banners de Redes Sociales,
            el estilo debe ser emocional (confianza y seguridad) de máximo 23 palabras, añadele emoticones"""
        }],
        max_tokens=70
    )

    Copy = respuesta.choices[0].message.content.strip()
    resultados.append(Copy)

df_productos["Copy"] = resultados
display(df_productos)
```
